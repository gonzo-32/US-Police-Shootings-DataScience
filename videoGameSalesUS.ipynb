```python
from urllib.request import urlretrieve
import pandas as pd
import matplotlib.pyplot as plt

URL = 'https://raw.githubusercontent.com/knpraveen/VideoGameSales/master/vgsales.csv'

urlretrieve(URL, 'vgsales.csv') 
```

```python
!head vgsales.csv
```

```python
DATA = pd.read_csv('vgsales.csv')
DATA.head(10)
```

```python
year_sales = DATA[["Year","NA_Sales"]]

year_sales.head()
```

```python
total_sales_per_year = year_sales.groupby(['Year'], sort = False).sum()
total_sales_per_year.head()

fig, tys = plt.subplots(figsize=(10,6))
xax = total_sales_per_year.index.values #Plotting the index in this case YEARS
yax = total_sales_per_year.NA_Sales
tys.bar(xax, yax)
```

```python
restructureDATA = DATA[["Platform", "Genre", "Publisher", "NA_Sales"]]
restructureDATA.head(10)
```

```python
x = restructureDATA.Platform
y = restructureDATA.NA_Sales

plt.figure(figsize=(10,10))
plt.scatter(x, y, marker='x')
plt.tick_params(axis='x', which='major', labelsize=13)
plt.xticks(rotation=(90))
plt.show()
```
