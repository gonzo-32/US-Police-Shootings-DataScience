```python
from urllib.request import urlretrieve
import pandas as pd
import matplotlib as mp

URL = 'https://github.com/washingtonpost/data-police-shootings/releases/download/v0.1/fatal-police-shootings-data.csv'

urlretrieve(URL, 'shootings.csv')
```

```python
!head shootings.csv
```

```python
DATA = pd.read_csv('shootings.csv', index_col='date', parse_dates=True)
DATA.head()
```

```python
%matplotlib inline
DATA.resample('W').sum().plot()
```
